# Imputing for Data_application

```python
# We will impute for  'OCCUPATION_TYPE' as it has null value percentage
inp0['OCCUPATION_TYPE'].fillna(value='missing', inplace= True)
```

```python
#Ext_Source3 and Target, it has 19.8% nulls, hence impute with median
inp0['EXT_SOURCE_3'].fillna(inp0['EXT_SOURCE_3'].median(), inplace= True)
```

```python
#Imputing columns as all are number of credit inquiries,impute it with 13% value using mode().
inp0['AMT_REQ_CREDIT_BUREAU_DAY'].fillna(
    value=inp0['AMT_REQ_CREDIT_BUREAU_DAY'].mode()[0],
    inplace=True)
inp0['AMT_REQ_CREDIT_BUREAU_YEAR'].fillna(
    value=inp0['AMT_REQ_CREDIT_BUREAU_YEAR'].mode()[0],
    inplace=True)
inp0['AMT_REQ_CREDIT_BUREAU_MON'].fillna(
    value=inp0['AMT_REQ_CREDIT_BUREAU_MON'].mode()[0],
    inplace=True)
inp0['AMT_REQ_CREDIT_BUREAU_WEEK'].fillna(
    value=inp0['AMT_REQ_CREDIT_BUREAU_WEEK'].mode()[0],
    inplace=True)
inp0['AMT_REQ_CREDIT_BUREAU_HOUR'].fillna(
    value=inp0['AMT_REQ_CREDIT_BUREAU_HOUR'].mode()[0],
    inplace=True)
inp0['AMT_REQ_CREDIT_BUREAU_QRT'].fillna(
    value=inp0['AMT_REQ_CREDIT_BUREAU_QRT'].mode()[0],
    inplace=True)
```

```python
# object_category has .42% missing value, hence impute using mode()
inp0['NAME_TYPE_SUITE'].fillna(inp0['NAME_TYPE_SUITE'].mode()[0], inplace= True)
```

```python
#impute data for columns less than 1% of null values using median
inp0['OBS_30_CNT_SOCIAL_CIRCLE'].fillna(
    value=inp0['OBS_30_CNT_SOCIAL_CIRCLE'].median(), inplace=True)
inp0['DEF_30_CNT_SOCIAL_CIRCLE'].fillna(
    value=inp0['DEF_30_CNT_SOCIAL_CIRCLE'].median(), inplace=True)
inp0['OBS_60_CNT_SOCIAL_CIRCLE'].fillna(
    value=inp0['OBS_60_CNT_SOCIAL_CIRCLE'].median(), inplace=True)
inp0['DEF_60_CNT_SOCIAL_CIRCLE'].fillna(
    value=inp0['DEF_60_CNT_SOCIAL_CIRCLE'].median(), inplace=True)
inp0['EXT_SOURCE_2'].fillna(
    value=inp0['EXT_SOURCE_2'].median(), inplace=True)
inp0['AMT_GOODS_PRICE'].fillna(
    value=inp0['AMT_GOODS_PRICE'].median(), inplace=True)
inp0['AMT_ANNUITY'].fillna(
    value=inp0['AMT_ANNUITY'].median(), inplace=True)
inp0['CNT_FAM_MEMBERS'].fillna(
    value=inp0['CNT_FAM_MEMBERS'].median(), inplace=True)
inp0['DAYS_LAST_PHONE_CHANGE'].fillna(
    value=inp0['DAYS_LAST_PHONE_CHANGE'].median(), inplace=True)
```
